In order to see the effect of some of the network analysis tools, it is useful to be able to generate graphs of a particular shape. There are many Graph Generators built into networkx. Some of the most common graph generators are illustrated below.

# Clasic Graphs

## empty_graph([n,Â create_using])

Return the empty graph with n nodes and zero edges.

```python
# Import necessary modules
import networkx as nx
import matplotlib.pyplot as plt
    
G = nx.empty_graph(n=10)

nx.draw(G, with_labels=True)
plt.show()
```

## cycle_graph(n[, create_using])

Return the cycle graph C_n over n nodes.

```python
# Import necessary modules
import matplotlib.pyplot as plt
import networkx as nx 

G = nx.cycle_graph(n=15)

nx.draw(G, with_labels=True)
plt.show()
```

## star_graph(n[, create_using])

Return the Star graph with n+1 nodes: one center node, connected to n outer nodes.

```python
# Import necessary modules
import matplotlib.pyplot as plt
import networkx as nx 

G = nx.star_graph(n=10)

nx.draw(G, with_labels=True)
plt.show()
```

## wheel_graph(n[, create_using])

Return the wheel graph: a single hub node connected to each node of the (n-1)-node cycle graph.

```python
# Import necessary modules
import matplotlib.pyplot as plt
import networkx as nx 

G = nx.wheel_graph(n=10)

nx.draw(G, with_labels=True)
plt.show()
```

## ladder_graph(n[, create_using])

Return the Ladder graph of length n.

```python
# Import necessary modules
import matplotlib.pyplot as plt
import networkx as nx 

G = nx.ladder_graph(n=6)

nx.draw(G, with_labels=True)
plt.show()
```

## lollipop_graph(m, n[, create_using])

## barbell_graph(m1,Â m2[,Â create_using])

## erdos_renyi_graph(n, p[, seed, directed])

Return a random graph G_{n,p} (ErdÅs-RÃ©nyi graph, binomial graph).

```python
# Import necessary modules
import matplotlib.pyplot as plt
import networkx as nx 

G = nx.erdos_renyi_graph(n=20, p=0.2)

nx.draw(G, with_labels=True)
plt.show()
```

```python
random_shell_graph(constructor[, seed])
```

```python
# Import necessary modules
import matplotlib.pyplot as plt
import networkx as nx 

# constructor (a list of three-tuples) â (n,m,d) for each shell starting at the center shell.
# n (int) â The number of nodes in the shell
# m (int) â The number or edges in the shell
# d (float) â The ratio of inter-shell (next) edges to intra-shell edges. d=0 means no intra shell edges, d=1 for the last shell

constructor=[(10,20,0.8)]

G = nx.random_shell_graph(constructor)

nx.draw(G, with_labels=True)
plt.show()
```

## random_clustered_graph(joint_degree_sequence)

Generate a random graph with the given joint degree and triangle degree sequence.

```python
# Import necessary modules
import matplotlib.pyplot as plt
import networkx as nx 

deg_tri=[[1,0],[1,0],[1,0],[2,0],[1,0],[2,1],[0,1],[0,1]]
G = nx.random_clustered_graph(deg_tri)
# G.remove_edges_from(G.selfloop_edges())

nx.draw(G, with_labels=True)
plt.show()
```
