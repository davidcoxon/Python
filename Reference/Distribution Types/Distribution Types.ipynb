# Distribution Types

You can simulate various types of probability distribution using Numpy by generating a large sample of random numbers with defined charachteristics, mean, standard deviation sample size etc. The more samples you do the longer it will take but the more precise the model distribution will be.

## Normal (or Gaussian) Distribution

This is the most common type of distribution, it is generated based on a mean and standard deviation, some time called mu μ (mean) & sigma Σ or σ (standard deviation)

>**np.random.normal(mean, standard diviation, size=sample size)**

example *np.random.normal(20, 1.5, size=10000)*

```python
# Normal Distribution

import numpy as np
import matplotlib.pyplot as plt
import pylab

# Draw 100000 samples from Normal distribution with stds of interest: samples_std1, samples_std3, samples_std10

samples_std1=np.random.normal(20,1,size=100000)
samples_std3=np.random.normal(20,3,size=100000)
samples_std10=np.random.normal(20,10,size=100000)

# Make histograms
_=plt.hist(samples_std1,normed=True, bins=100,color='blue',histtype='step') # histtype step remove the fill below the line
_=plt.hist(samples_std3,normed=True, bins=100, color='red',histtype='step')
_=plt.hist(samples_std10,normed=True, bins=100, color='green', histtype='step')
 

# Make a legend, set limits and show plot
_ = plt.legend(('std = 1', 'std = 3', 'std = 10'))
_ = plt.title('Normal Distributions (with different std var)')
plt.ylim(-0.01, 0.42)

pylab.savefig('Graphs/normaldist.png')   # save the figure to file
pylab.savefig('Graphs/normaldist.pdf') # savefig('Graphs/normaldistribution.pdf')
plt.show()
```

## Bernoulli Trials

A Bernoulli Trial is an experiment that has two options,"success" (True) and "failure" (False).

## Binomial distribution

The number r of successes in n Bernoulli trials with probability p of success, is described as a Binomially distributed.

Example: *The number of heads in 4 coin flips with probability 0.5 of heads*

>**np.random.binomial(number of tials, probability of success,size=sample size)**

Example: *np.random.binomial(60, 0.5, size=100)* #creates an array of 60 Bernoulli Trials a binomial distribution and a probability of success of 0.5, repeated 100 times.

Note: the Biniomial is based on Bernoulli Trials, making it discrete so unlike a continuous distribution it will be a histogram rather that a smooth curve.

# Poisson Distributions

A Poisson Distribution is the number r of arrivals of a Poisson process in a given time interval with average rate of λ arrivals per interval

Example: The number r of hits on a website in one hour with an average hit rate of 6 hits per hour is Poisson distributed.

The Poisson Disstribution is a limit of the Binomial distribution for low probability of sucess and a large number of Bernoulli trials (ie for rare events).

```python
# Poisson Distribution

import numpy as np
import matplotlib.pyplot as plt
import pylab

poisson = np.random.poisson(6, size=10000)
_ = plt.hist(poisson, bins=100)
plt.margins(0.02)
_ = plt.title('Poisson Distribution')
_ = plt.xlabel('number of successes')
_ = plt.ylabel('distribution')

pylab.savefig('Graphs/poissonDistribution.png')   # save the figure to file
pylab.savefig('Graphs/poissonDistribution.pdf') # savefig('Graphs/exponentialcdf.pdf')
plt.show()
```

## The Exponential distribution

With an Exponential distribution the waiting time between arrivals of a Poisson process is Exponentially distributed.

>**np.random.exponential(tau, size)**

Example: *np.random.exponential(175, size=10000)

Tau (Τ or τ\) is the rate of decay in the poisson process. The lower the value the greater the higher the starting point and the sleeper the slope of the distribution. 

```python
# Exponential Distribution

import numpy as np
import matplotlib.pyplot as plt
import pylab

# Draw 100000 samples from exponential distribution with stds of interest: samples_std1, samples_std3, samples_std10

samples_tau1=np.random.exponential(50,size=100000)
samples_tau2=np.random.exponential(150,size=100000)
samples_tau3=np.random.exponential(250,size=100000)

# Make histograms
_=plt.hist(samples_tau1,normed=True, bins=100,color='blue',histtype='step')
_=plt.hist(samples_tau2,normed=True, bins=100, color='red',histtype='step')
_=plt.hist(samples_tau3,normed=True, bins=100, color='green', histtype='step')
 

# Make a legend, set limits and show plot
_ = plt.legend(('tau = 50', 'tau = 150', 'tau = 250'))
_ = plt.title('Exponential Distributions (with different tau)')

pylab.savefig('Graphs/exponentialdist.png')   # save the figure to file
pylab.savefig('Graphs/exponentialdist.pdf') # savefig('Graphs/normaldistribution.pdf')
plt.show()
```

# The Cumulative distribution Function (CDF)

The Cumulative distribution function for each distribution shows the cumulative totals, this distribution avoids problems caused by binning.  

## Normal CDF

Each of the Cumulative distribution functions are calculated from there distribution functions using an ecdf function.

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x=np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, len(x)+1) / len(x)

    return x, y
```

# Comparing Cumulative Distribution functions
