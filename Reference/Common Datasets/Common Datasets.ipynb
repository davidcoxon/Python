# Sample Datasets

Within Datascience there are a number of commonly used datasets. These typically used for training and demonstration purposes. One of the reason's that these particular datasets maybe so popular for teaching purposes is that they contain data that most people with and able to connect with and will find reasonably familiar. 

For example:  there is a dataset on weather in warsaw, while not everyone will have been to Warsaw its lilkely that most people will be familiar with seasonal changes within the data and could anticipate variations in temperature and rainfall between summer and winter and may not be surprised to find some outliers within the general pattern or a degree of randomness in day to day results. There is another data set on mileage per gallon for various models of car, its likely that most people will be familiar enough with cars to make a resonable guess that the bigger the car engine the more displacement it will have and the greater power the engine will have but the fewer miles per gallon it will be able to achieve, the dataset also contains details of car weights , accelerations , years and numbers of cyclinders. This mpg dataset has a lot more variable than the weather data so it can be used to demonstrate more complex multivariant relationships. Other datasets such as the iris data show how 4 simple measures petal and sepal length and width vary in 3 different varieties of iris, while the Hans roslings Gapminder dataset contains a single annual figure for 780 countries take over 218 years. Otherdata sets like the spring data contain multile categories of data like the countries athletes came from the the category of medal won along side their names and sprint times.

There is pretty much a samples dataset available to download for every situation you might want to demonstrate. Having a good understanding of each of these dataset and how the different types of exploratory data analysis works will equip you with the ability to be able to use these datasets to explore new libraies and functions of your own. In this notebook i have explored a selection of the datasets that came with the bokeh library (a library for producing interactive visualizations). For each dataset there is a basis statistical and graphical exploratory analysis. For some datasets where there appears to be an interesting relationship or pattern there is some additional analsis.

I would encourage any potential datascience students to create their own note book and conduct their own exploratory analysis of these common dataset. 

# aapl dataset

```python
# import aapl dataset

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves
import pylab

# Get aapl stock Data
filename="Data/aapl.csv"
aapl=pd.read_csv(filename,encoding = "ISO-8859-1",parse_dates=True, index_col='date', sep=',')

# statistical EDA

print("AAPL dataset head()\n")
print(aapl.head())

print("\nAAPL dataset shape\n")
print(aapl.shape)
      
print("\nAAPL dataset info()\n")
print(aapl.info())

print("\nAAPL dataset describe()\n")
print(aapl.describe())

# Graphical EDA

plt.figure(1)
aapl.plot(x='open', y='close',kind='scatter',color='blue')
plt.figure(2)
aapl.plot(x='high',y='low',kind='scatter', color="red")
plt.figure(3)
aapl.plot(x='high',y='volume',kind='scatter', color="green")

plt.show()
```

# AUTO-MPG DATASET

```python
# import auto-mpg data

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves
import pylab

# Get Data
filename="Data/auto-mpg.csv"
auto=pd.read_csv(filename, sep=',')

# statistical EDA

print("Auto dataset head()\n")
print(auto.head())

print("\nAuto dataset shape\n")
print(auto.shape)

print("\nAuto dataset info()\n")
print(auto.info())

print("\nAuto dataset describe()\n")
print(auto.describe())

# Graphical EDA

plt.figure(1)
auto.plot(x='mpg', y='hp', kind='scatter', alpha=0.5, color='red')
plt.figure(2)
auto.plot(x='mpg', y='displ', kind='scatter', alpha=0.5, color='blue')
plt.figure(3)
auto.plot(x='mpg', y='accel', kind='scatter', alpha=0.5,color='green')
plt.figure(4)
auto.plot(x='mpg', y='weight', kind='scatter', alpha=0.5,color='purple')
plt.figure(5)
auto.plot(x='displ', y='weight', kind='scatter', alpha=0.5,color='purple')

# Plot the PDF
plt.figure(6)
auto.displ.plot(kind='hist',normed=True, bins=20, alpha=0.5,color='blue')
plt.figure(7)
auto.accel.plot(kind='hist',normed=True, bins=20, alpha=0.3,color='green')
plt.figure(8)
auto.mpg.plot(kind='hist',normed=True, bins=20, alpha=0.3,color='red')
plt.figure(9)
auto.displ.plot(kind='hist',normed=True, bins=20, alpha=0.3,color='blue')
plt.show()
```

# iris dataset

```python
# iris dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
import pylab

# get data
from bokeh.sampledata.iris import flowers as iris

# Statistical eda

print("Iris dataset head()\n")
print(iris.head())
print("\nIris dataset shape\n")
print(iris.shape)
print("\nIris dataset info()\n")
print(iris.info())
print("\nIris dataset describe()\n")
print(iris.describe())
print("\nIris dataset species column describe()\n")
print(iris['species'].describe())
print("\nIris dataset species column unique()\n")
print(iris['species'].unique())

# Plot the PDF
plt.figure(1)
(iris[iris['species']=='versicolor']).plot(kind='hist',normed=True,range=(0,8), bins=50, alpha=0.3)

# Plot the CDF
plt.figure(2)
(iris[iris['species']=='versicolor']).plot(kind='hist', normed=True, bins=50, cumulative=True, alpha=0.3)

plt.show()
```
