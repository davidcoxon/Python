# Noughts and Crosses

‘Noughts and Crosses’ is a classic kids game. The rules are simple and only only takes seconds to play each game, so it's the idea candidate for learning to code.

The original concept for this project came from a CodeClub project which has since been taken on the codeClub site.

## Drawing the grid

The first step to playing and game of Noughts and Crosses is to set up the board. We are going to use Tkinter as our canvas. We'll import tkinter, inisiate an instance and define a canvas size of 600 x 600.

Next we'll create some lines (2 horizontal and 2 vertical) using create_line(). Create_line() draws a straight line between 2 points and takes 4 arguments separated by commas, the x and y co-ordinates on the start point and the x and y co-ordinates of the end point.

Finally we'll call the loop to display the canvas.

![noughtsandcrossesfinal](Images/grid1.png?raw+True)

```python
#import library
from tkinter import *

# create canvas
main = Tk()
c = Canvas(main, width=600, height=600)
c.pack()

# Create grid
c.create_line(200, 0, 200, 600)
c.create_line(400, 0, 400, 600)
c.create_line(0, 200, 600, 200)
c.create_line(0, 400, 600, 400)

# initiate loop
mainloop()
```

## Drawing an O

Once we have the board we will need to place x's and o's on to it. We'll be adding between 3 and 5 of each symbol on the board (depending on how quickly we get a winner) we will create a function to draw them.

To draw an O we'll use create_oval(), again this takes 4 arguments. If we drew a box around the oval these would be the bottom left and top right, x and y co-ordinates of that box.

We'll start with static co-ordinates 200,200 and 400,400 to draw the oval in the middle of the screen. 

We'll create the oval inside of a function named click(event) and we will use the .bind method to bind the click function to the button.

![noughtsandcrossesfinal](Images/o.png?raw+True)

```python
# import library
from tkinter import *

# create canvas
main = Tk()
c = Canvas(main, width=600, height=600)
c.pack()

# create a click event
def click(event):
    c.create_oval(200,200,400,400)
c.bind('<Button-1>', click)

# initiate loop
mainloop()
```

## Drawing X's and O's on the grid.

We already know how to draw an x, as an x is simply 2 lines in opposite directions that cross in the middle, so its the same technique as we used to created the grid.

We also know how to draw the O, which is just an oval.

Before we create our function to generate the x's and o's, we'll create a variable shape to keep track of whether we are creating an 'x' or 'o'. We'll use global shape within the function to access the variable.

We'll create our function to draw the X's and 0's which we'll use a number of times as we play the game. Instead of passing numbers for the co-ordinated of the circles and lines we'll use variables which we'll get these from the co-ordinates of the custor when clicked. We'll divide by 200 and force them to be int's and multiply them up by 200 again, this forces each symbol to be drawn in the right place on the grid.

If we're drawing an x once we've finished we'll change the shape varaible to ensure that the next sysmbol drawn will be an O, and when we've drawn that we'll change it again to ensure the next is an X, and so on.

```python
from tkinter import *
main = Tk()
c = Canvas(main, width=600, height=600)
c.pack()

shape = 'O'
def click(event):
    global shape
    across = int(c.canvasx(event.x)/200)
    down = int(c.canvasy(event.y)/200)
    if shape == 'O':
        c.create_oval(
            across*200,down*200,
            (across+1)*200,(down+1)*200
        )
        shape = 'X'
    else:
        c.create_line(
            across*200, down*200,
            (across+1)*200, (down+1)*200
            )
        c.create_line(
            across*200, (down+1)*200,
            (across+1)*200, down*200
            )
        shape = 'O'
c.bind('<Button-1>', click)
mainloop()
```

## Keeping track

Now that we have a board and can add X's and O's on it in the right places, we need to define a grid, each square on the grid will have a value of between 0 and 8, so that it can be called and we can assign it the value of either X or O.

```python
from tkinter import *
main = Tk()
c = Canvas(main, width=600, height=600)
c.pack()
c.create_line(200, 0, 200, 600)
c.create_line(400, 0, 400, 600)
c.create_line(0, 200, 600, 200)
c.create_line(0, 400, 600, 400)
shape = 'O'
grid = [
    '0', '1', '2',
    '3', '4', '5',
    '6', '7', '8',
    ]
def click(event):
    global shape, grid
    across = int(c.canvasx(event.x)/200)
    down = int(c.canvasy(event.y)/200)
    square = across + (down*3)
    if grid[square] == 'X' or grid[square] == 'O':
        return
    if shape == 'O':
        c.create_oval(
            across*200,down*200,
            (across+1)*200,(down+1)*200
        )
        grid[square] = 'O'
        shape = 'X'
    else:
        c.create_line(
            across*200, down*200,
            (across+1)*200, (down+1)*200
        )
        c.create_line(
            across*200, (down+1)*200,
            (across+1)*200, down*200
        )
        grid[square] = 'X'
        shape = 'O'
c.bind('<Button-1>', click)
mainloop()
```

## Finding a winner

Now we have a defined a board, draw some x's and o's and  started tracking the locations od each players symbols,we start evaluating the pieces to see when we have a line of three x's or O's. We'll be doing this after each new piece is placed, so we'll define a function to do this more efficiently. We'llcall this function winner. We'll use an if statement within our click function to detect when winner has been trigger.

```python
from tkinter import *
main = Tk()
c = Canvas(main, width=600, height=600)
c.pack()
c.create_line(200, 0, 200, 600)
c.create_line(400, 0, 400, 600)
c.create_line(0, 200, 600, 200)
c.create_line(0, 400, 600, 400)
shape = 'O'
grid = [
    '0', '1', '2',
    '3', '4', '5',
    '6', '7', '8',
    ]
def click(event):
    global shape, grid
    across = int(c.canvasx(event.x)/200)
    down = int(c.canvasy(event.y)/200)
    square = across + (down*3)
    if grid[square] == 'X' or grid[square] == 'O':
        return
    if winner():
        return
    
    if shape == 'O':
        c.create_oval(
            across*200,down*200,
            (across+1)*200,(down+1)*200
        )
        grid[square] = 'O'
        shape = 'X'
    else:
        c.create_line(
            across*200, down*200,
            (across+1)*200, (down+1)*200
        )
        c.create_line(
            across*200, (down+1)*200,
            (across+1)*200, down*200
        )
        grid[square] = 'X'
        shape = 'O'
        
def winner():
    for across in range(3):
        row = across*3
        line = grid[row] + grid[row+1] + grid[row+2]
        if line == 'XXX' or line == 'OOO':
            return True
    for down in range(3):
        line = grid[down] + grid[down+3] + grid[down+6]
        if line == 'XXX' or line == 'OOO':
            return True
    line = grid[0]+grid[4]+grid[8]
    if line == 'XXX' or line == 'OOO':
        return True
    line = grid[2]+grid[4]+grid[6]
    if line == 'XXX' or line == 'OOO':
        return True

c.bind('<Button-1>', click)
mainloop()
```

## What if there's no winner?

So if either player gets a row, we get a message saying who's won, but what if we get to the end of the game and neither player has won, we could just leave it as it is with a full board, but if we are going to print a message saying we have a winner when a player wins shouldn't we have a message when players draw. 

I through about a couple of ways to do this the, first involved initiating a count and increasing it each time a grid was asigned to X or 0 and declaring a draw if you got to 9 and winner had still not been declared.

I also thought about iterating over the grid after each move and checking each square to see if had been assigned either X or O to establish when the game was over.  

In ended up going with a function that loops over each square in the grid and looking for blanks, when all the squares in the grid are taken if there has not been a winner we have a draw. Because on click will only evaluate the results on the next click and you can only click on a blank square i had to change the action to on exit.  

```python
from tkinter import *
main = Tk()
c = Canvas(main, width=600, height=600)
c.pack()
c.create_line(200, 0, 200, 600)
c.create_line(400, 0, 400, 600)
c.create_line(0, 200, 600, 200)
c.create_line(0, 400, 600, 400)
shape = 'O'
grid = [
    '0', '1', '2',
    '3', '4', '5',
    '6', '7', '8',
    ]
def click(event):
    global shape, grid
    across = int(c.canvasx(event.x)/200)
    down = int(c.canvasy(event.y)/200)
    square = across + (down*3)
    if grid[square] == 'X' or grid[square] == 'O':
        return
    if winner():
        def onexit():
           pass

        text = Text(main)
        if shape == 'O':
            text.insert(INSERT, "X's win!")
        else:
            text.insert(INSERT, "O's win!")
        
        text.pack()
        text.tag_add("here", "1.0", "2.4")
        text.tag_config("here", background="white", foreground="red",font='helvetica 18 bold', justify='center')

        main.mainloop()
    if draw():
        def onexit():
           pass

        text = Text(main)
        text.insert(INSERT, "We have a draw!")
        
        text.pack()
        text.tag_add("here", "1.0", "2.4")
        text.tag_config("here", background="white", foreground="red",font='helvetica 18 bold', justify='center')

        main.mainloop()
        
        return
    if shape == 'O':
        c.create_oval(
            across*200,down*200,
            (across+1)*200,(down+1)*200
        )
        grid[square] = 'O'
        shape = 'X'
    else:
        c.create_line(
            across*200, down*200,
            (across+1)*200, (down+1)*200
        )
        c.create_line(
            across*200, (down+1)*200,
            (across+1)*200, down*200
        )
        grid[square] = 'X'
        shape = 'O'       
        
def winner():
    for across in range(3):
        row = across*3
        line = grid[row] + grid[row+1] + grid[row+2]
        if line == 'XXX' or line == 'OOO':
            return True
    for down in range(3):
        line = grid[down] + grid[down+3] + grid[down+6]
        if line == 'XXX' or line == 'OOO':
            return True
    line = grid[0]+grid[4]+grid[8]
    if line == 'XXX' or line == 'OOO':
        return True
    line = grid[2]+grid[4]+grid[6]
    if line == 'XXX' or line == 'OOO':
        return True
    
def draw():
    blanks=0
    for square in range(9):
        if grid[square] != 'X' and grid[square] != 'O':
           blanks=blanks+1
        
    if blanks ==1:
        return True
    

c.bind('<Button-1>', click)
mainloop()
```
